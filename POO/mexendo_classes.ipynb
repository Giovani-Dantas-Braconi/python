```python
from datetime import datetime
import pytz

class Conta_corrente:
    @staticmethod
    def _data_hora():
        fuso_BR = pytz.timezone('Brazil/East')
        horario_BR = datetime.now(fuso_BR)
        return horario_BR.strftime("%d/%m/%y %H:%M:%S")

    def __init__(self, saldo: float):
        self.saldo = saldo
        self.limite = -1000
        self.transacoes = []

    def verify(self, dindin: float):
        jogada = self.saldo - dindin
        if jogada >= self.limite:
            return True
        else:
            return False

    def depositar(self, dindin: float):
        self.saldo += dindin
        self.transacoes.append((dindin, self.saldo, Conta_corrente._data_hora()))
        return f"entrou {dindin} e o seu saldo é de {self.saldo}"
    
    def sacar(self, dindin: float):
        if self.verify(dindin):
            self.saldo -= dindin
            self.transacoes.append((-dindin, self.saldo, Conta_corrente._data_hora()))
            return f"saiu {dindin} e o seu saldo é de {self.saldo}"
        else:
            return f"o seu limite não permite você sacar mais que isso"

    def verificador_deb(self):
        return f"o seu saldo é de {self.saldo}"
    

    def consutar_historico_transacoes(self):
        print("Histórico das trancções: ")
        print("Valor", "Saldo", "Data e Hora")
        for transacoes in self.transacoes:
            print(transacoes)

melo = Conta_corrente(400)
print(melo.depositar(100))
print(melo.sacar(0))
print(melo.verificador_deb())
print(melo.sacar(150))
print(melo.verificador_deb())
print(melo.transacoes)
print("\n\n\n\n\n\n\n\n\n\n\n\n")
melo.consutar_historico_transacoes()
```

# MSM CODIGO COM DOCSTRING, ESTRANHO PRA KRL 

```python
class Conta1:
    def __init__(self):
        self.saldo = 1100

    def transferir(self, qntd:float, conta):
        """
        Transfere uma certa quantidade de dinheiro desta conta para outra conta.

        :param qntd: A quantidade de dinheiro a ser transferida.
        :param conta: A conta de destino para a transferência.
        :return: O saldo atual desta conta após a transferência.
        """
        conta.saldo -= qntd
        self.saldo += qntd
        return self.saldo

    def depositar(self, qntd:float, conta):
        """
        Deposita uma certa quantidade de dinheiro nesta conta vinda de outra conta.

        :param qntd: A quantidade de dinheiro a ser depositada.
        :param conta: A conta de origem do depósito.
        :return: O saldo atual desta conta após o depósito.
        """
        conta.saldo += qntd
        self.saldo -= qntd
        return self.saldo

class Conta2:
    def __init__(self):
        self.saldo = 3000

    def transferir(self, qntd:float, conta):
        """
        Transfere uma certa quantidade de dinheiro desta conta para outra conta.

        :param qntd: A quantidade de dinheiro a ser transferida.
        :param conta: A conta de destino para a transferência.
        :return: O saldo atual desta conta após a transferência.
        """
        conta.saldo -= qntd
        self.saldo += qntd
        return self.saldo

    def depositar(self, qntd:float, conta):
        """
        Deposita uma certa quantidade de dinheiro nesta conta vinda de outra conta.

        :param qntd: A quantidade de dinheiro a ser depositada.
        :param conta: A conta de origem do depósito.
        :return: O saldo atual desta conta após o depósito.
        """
        conta.saldo += qntd
        self.saldo -= qntd
        return self.saldo

def exibir_docstring(obj, method_name):
    """
    Exibe a docstring de um método específico de um objeto.

    :param obj: O objeto que contém o método.
    :param method_name: O nome do método.
    """
    method = getattr(obj, method_name)
    print(f"Docstring para '{method_name}':\n{method.__doc__}")

visu1 = Conta1()
visu2 = Conta2()

print(visu1.transferir(200, visu2))
print(visu2.saldo)
print("\n\n")
print(visu2.transferir(100, visu1))
print(visu1.saldo)

exibir_docstring(visu1, 'transferir')
exibir_docstring(visu1, 'depositar')
exibir_docstring(visu2, 'transferir')
exibir_docstring(visu2, 'depositar')
```
